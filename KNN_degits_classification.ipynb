```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

# Load the digits dataset
digits = load_digits()

# Create a DataFrame from the digits data
df = pd.DataFrame(data=digits.data)
df['target'] = digits.target  # Add target labels

# Set up the plot
plt.figure(figsize=(10, 8))

for i in range(10):
    df_digit = df[df.target == i] 
    sample = df_digit.sample(5, random_state=42) 

    for j, row in enumerate(sample.iterrows()):
        plt.subplot(10, 5, i * 5 + j + 1) 
        plt.imshow(row[1].values[:-1].reshape(8, 8), cmap='gray') 
        plt.title(f"Label: {i}")
        plt.axis('off') 

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
    X =  df.drop(['target'],axis = 'columns')
```

```python
y = df.target
```

```python
X_train , X_test ,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=41)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm,annot = True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
